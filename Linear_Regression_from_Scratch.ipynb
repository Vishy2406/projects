# Linear Regression from Scratch in Python

This project demonstrates how to build and visualize a simple linear regression model using only base Python and NumPy. We simulate a dataset of study hours vs. exam scores and compute the regression line manually.


## Simulate Dataset

We simulate 50 data points representing "Hours Studied" and corresponding "Exam Scores". We use a linear function with added noise to make it realistic:
- Underlying model: **y = 5x + 40 + noise**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate data: Study Hours vs. Exam Score
np.random.seed(0)
hours = np.random.uniform(1, 10, 50)
noise = np.random.normal(0, 5, 50)
scores = 5 * hours + 40 + noise  # y = 5x + 40 + noise

# Store in DataFrame
data = pd.DataFrame({
    'Hours_Studied': hours,
    'Exam_Score': scores
})

# Plot it
plt.scatter(data['Hours_Studied'], data['Exam_Score'])
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Simulated Study Data")
plt.grid(True)
plt.show()
```

## Linear Regression Logic

Instead of using scikit-learn, we calculate the regression line manually using these formulas:

- **Slope (m)** = Σ(x - x̄)(y - ȳ) / Σ(x - x̄)²  
- **Intercept (b)** = ȳ - m·x̄

This gives us the best-fit line: **y = mx + b**


```python
# Calculate means
x = data['Hours_Studied']
y = data['Exam_Score']
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate slope (m) and intercept (b)
numerator = sum((x - x_mean) * (y - y_mean))
denominator = sum((x - x_mean)**2)
m = numerator / denominator
b = y_mean - m * x_mean

print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
```

## Visualize Regression Line

We plot the original data points and overlay the regression line we computed manually.


```python
# Calculate means
x = data['Hours_Studied']
y = data['Exam_Score']
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate slope (m) and intercept (b)
numerator = sum((x - x_mean) * (y - y_mean))
denominator = sum((x - x_mean)**2)
m = numerator / denominator
b = y_mean - m * x_mean

# Prediction line
y_pred = m * x + b

# Plot
plt.figure(figsize=(8,6))
plt.scatter(x, y, label="Data Points")
plt.plot(x, y_pred, color='red', label="Regression Line")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Linear Regression From Scratch")
plt.legend()
plt.grid(True)
plt.show()
```

## Evaluate Model (Mean Squared Error)

We calculate the Mean Squared Error(MSE) between the actual and predicted exam scores:
**MSE = (1/n) · Σ(actual - predicted)²**

A lower MSE means better prediction accuracy.

```python
# Mean Squared Error
mse = np.mean((y - y_pred) ** 2)
print(f"Mean Squared Error: {mse:.2f}")
```
